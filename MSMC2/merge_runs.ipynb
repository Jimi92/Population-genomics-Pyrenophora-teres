```python
import glob
import pandas as pd
import numpy as np

Mor_files = glob.glob("Morocco_*final.txt")
Mor_files
```

```python
df_1 = pd.read_csv("Morocco_1.msmc.final.txt", sep="\t")
df_2 = pd.read_csv("Morocco_2.msmc.final.txt", sep="\t")
df_3 = pd.read_csv("Morocco_3.msmc.final.txt", sep="\t")
df_4 = pd.read_csv("Morocco_4.msmc.final.txt", sep="\t")
df_5 = pd.read_csv("Morocco_5.msmc.final.txt", sep="\t")
df_1
```

```python
dfs = [df_1,df_2,df_3,df_4,df_5]
```

```python
result_1 = pd.merge(df_1,df_2, how="outer", on=["time_index","left_time_boundary","right_time_boundary"])
result_2 = pd.merge(result_1,df_3, how="outer", on=["time_index","left_time_boundary","right_time_boundary"])
result_3 = pd.merge(result_2,df_4, how="outer", on=["time_index","left_time_boundary","right_time_boundary"])
result_4 = pd.merge(result_3,df_5, how="outer", on=["time_index","left_time_boundary","right_time_boundary"])


sorted_df = result_4.sort_values("left_time_boundary")

sorted_fill = sorted_df.interpolate()
full_fill_Mor = sorted_fill.bfill()
full_fill_Mor.to_csv("Full_table_Mor", sep = "\t", index = False)
```
