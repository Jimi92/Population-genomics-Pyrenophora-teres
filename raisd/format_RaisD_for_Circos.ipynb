```python
import pandas as pd
import numpy as np
import glob
```

```python
col_names=['Pos', 'Start', 'End', 'Pi', 'SFS','LD','Mu'] 
def read_RaisD_Reps(f,chrom):
    df = pd.read_table(f, skiprows= 1, names=col_names)
    df['Chr'] = chrom
    df = df[['Chr','Start','End','Mu']]
    return df
```

```python
chr1 = read_RaisD_Reps('RAiSD_Report.France_Chr1.out','Chr1')
chr2 = read_RaisD_Reps('RAiSD_Report.France_Chr2.out','Chr2')
chr3 = read_RaisD_Reps('RAiSD_Report.France_Chr3.out','Chr3')
chr4 = read_RaisD_Reps('RAiSD_Report.France_Chr4.out','Chr4')
chr5 = read_RaisD_Reps('RAiSD_Report.France_Chr5.out','Chr5')
chr6 = read_RaisD_Reps('RAiSD_Report.France_Chr6.out','Chr6')
chr7 = read_RaisD_Reps('RAiSD_Report.France_Chr7.out','Chr7')
chr8 = read_RaisD_Reps('RAiSD_Report.France_Chr8.out','Chr8')
chr9 = read_RaisD_Reps('RAiSD_Report.France_Chr9.out','Chr9')
chr10 = read_RaisD_Reps('RAiSD_Report.France_Chr10.out','Chr10')
chr11 = read_RaisD_Reps('RAiSD_Report.France_Chr11.out','Chr11')
chr12 = read_RaisD_Reps('RAiSD_Report.France_Chr12.out','Chr12')
```
