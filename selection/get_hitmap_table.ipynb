```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_ss_gene_overlap_table(f):
    df = pd.read_table(f, header = 0)
    df["Presence"] = 1
    return df
```

```python
c1 = read_ss_gene_overlap_table("C1_genes/C1_ss_gene_overlap.bed")
c2 = read_ss_gene_overlap_table("C2_genes/C2_ss_gene_overlap.bed")
c3 = read_ss_gene_overlap_table("C3_genes/C3_ss_gene_overlap.bed")
c4 = read_ss_gene_overlap_table("C4_genes/C4_ss_gene_overlap.bed")
c5 = read_ss_gene_overlap_table("C5_genes/C5_ss_gene_overlap.bed")
c6 = read_ss_gene_overlap_table("C6_genes/C6_ss_gene_overlap.bed")
```

```python
hm_df = pd.merge(c1,c2,on=["Chr","Start","End"], how='outer',suffixes=("_Ir", "_ND")).sort_values(by = "Chr")
hm_df = pd.merge(hm_df,c3,on=["Chr","Start","End"],how='outer',suffixes=("_1","_c3")).sort_values(by = "Chr")
hm_df = pd.merge(hm_df,c4,on=["Chr","Start","End"],how='outer',suffixes=("_Mor","_Fr")).sort_values(by = "Chr")
hm_df = pd.merge(hm_df,c5,on=["Chr","Start","End"],how='outer',suffixes=("_3","_c4")).sort_values(by = "Chr")
hm_df = pd.merge(hm_df,c6,on=["Chr","Start","End"],how='outer',suffixes=("_Az","_Wild")).sort_values(by = "Chr").reset_index(drop = True).fillna(0)
hm_df["score1"] = hm_df.Presence_Ir + hm_df.Presence_ND + hm_df.Presence_Mor + hm_df.Presence_Fr + hm_df.Presence_Az + hm_df.Presence_Wild
hm = hm_df.sort_values(by = 'score1', ascending=False)
hm = hm.iloc[:, 0:-1]
hm.to_csv('heatmap_table.txx', sep = '\t',index = False, header = True)
#hm_df.sort_values(by = ['Presence_Ir','Presence_ND','Presence_Mor','Presence_Fr','Presence_Az','Presence_Wild']).reset_index(drop = True)
```
